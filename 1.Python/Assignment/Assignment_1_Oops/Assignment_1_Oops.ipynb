### 1. Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.

```python
# Make bank class
class bank:
    """
    A mini banking system with deposit and withdraw feature.
    """
    log.info("Start of Banking System....")

    # Creating three attributes
    def __init__(self, accnum, name, balance=0):
        self.account_num = accnum
        self.holder_name = name
        self.balance = balance

    # Implement of deposit method
    def deposit(self, amount):
        """
        Function for deposit amount.

        Parameter:
        amount (int or float): The amount to be deposited.

        Return:
        int or float: The Total amount in the account of account holder.
        """
        try:
            if amount > 0:
                self.balance += amount
                log.info(f"Amount {amount} deposit successfully. Depositer current balance {self.balance}.")

                print(f"Amount {amount} deposit successfully. Your current balance {self.balance}.")
            else:
                log.error("Typed amount is less than 1.")
                raise ValueError("Deposit amount should be greater than 0.")
        
        except ValueError as v:
            print(v)
    
    # Implement of withdraw method
    def withdraw(self, amount):
        """
        Function for withdraw amount.

        Parameter:
        amount (int or float): The amount to be withdrawl.

        Return:
        int or float: The Total amount in the account of account holder.
        """
        try:
            if amount > 0:
                if self.balance >= amount:
                    self.balance -= amount
                    log.info(f"Take {amount}. Current balance is {self.balance}.")
                    
                    print(f"Take your amount {amount}. Your current balance {self.balance}.")
                
                else:
                    log.error("Typed amount greater than available balance.")
                    raise ValueError("Insufficient funds.")
            else:
                log.error("Typed amount is less than 1.")
                raise ValueError("Withdraw amount must be greater than 0.")
        
        except ValueError as v:
            print(v)


# Do an example usage:
# Create a bank account
acc = bank("123456789", "Rajdip Das", 10000)

# Deposit money
acc.deposit(500)

# Withdraw money
acc.withdraw(2000)  # This will log an error due to insufficient funds

# Another deposit
acc.deposit(100)
```

- Make a class named `bank`.
- Make 3 attribute named `account_num`, `holder_name`, `balance`.
- Make `deposit` and `withdraw` methods for perform such operations.
- Do all kinds of logging and exception handling in the class.
- Create a bank account with calling the class `bank` and pass arguments like, account number, name, and initial balance. Store it in `acc` object.
- Perform deposit and withdraw operations calling such methods using the object and print the deposit/withdraw balance and current balance.

### 2. Employee Management Create a class representing an employee with attributes like employee ID, name, and salary. Implement methods to calculate the yearly bonus and display employee details.

```python
# Create a employee class
class employee:
    """
    A employee management system. 
    
    Calculate yearly salary and display employee details.
    """
    log.info("Start of employee management system...")

    # Implement three attributes like emp_id, name and salary
    def __init__(self, emp_id, name, salary):
        self.emp_id = emp_id
        self.name = name
        self.salary = salary

    # Implement a method for calculate yearly bonus
    def calculate_yearly_bonus(self, percent):
        """
        Calculate yearly salary of the employee.

        Parameter:
        percent (int or float): How much yearly bonus percentage the employee got.

        Return:
        int or float: How much bonus amount the employee got.
        """
        try:
            if percent >= 0:
                self.bonus = (percent / 100) * self.salary
                log.info(f"Calculated yearly bonus for {self.salary}: {self.bonus}")
                return self.bonus
        
            else:
                log.error("Entered bonus percentage should be a positive number.")
                raise ValueError("Bonus percentage should be a positive number.")
            
        except ValueError as v:
            print(v)

    # Implement a method for display a employee details
    def display(self):
        """
        Display all the employee details.

        Parameter:
        None.

        Return:
        string: All details of an employee.
        """
        log.info("Show the employee details...")
        log.info(f"Employee ID: {self.emp_id}, Name: {self.name}, Salary: {self.salary}, bonus: {self.bonus}")
        print(f"Employee ID: {self.emp_id}, Name: {self.name}, Salary: {self.salary}, bonus: {self.bonus}")


# Do an example usage:
# Create an employee
emp = employee(1234, "Rajdip Das", 40000)

# Calculate yearly bonus and display it
bonus_amount = emp.calculate_yearly_bonus(20)
print("The bonus amount:", bonus_amount)

# Display employee details
emp.display()
```

### 3. Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle, return a vehicle, and display available vehicles.

```python
# Create a vehicleRental class
class vehicleRental:
    """
    A simple vehicle rental system.
    """
    log.info("Starting vehicle rental system....")

    # Implement an attribute called available_vehicle
    def __init__(self):
        self.available_vehicle = set()

    # Implement a method for add vehicle in the system
    def add_vehicle(self, vehicle):
        """
        Add a vehicle in the system.

        Parameter:
        vehicle (string): Vehicle name which you want to added in system.

        Return:
        string: A message vehicle added successfully in the system.
        """
        self.available_vehicle.add(vehicle)
        print(f"{vehicle} added successfully.")
        log.info(f"{vehicle} added to available vehicles.")

    # Implement a method for rent a vehicle from the system
    def rent_vehicle(self, vehicle):
        """
        Rent a vehicle from the system.

        Parameter:
        vehicle (string): Vehicle name which you want to rent from the system.

        Return:
        string: A message vehicle rent successfully or not.
        """
        try:
            if vehicle in self.available_vehicle:
                self.available_vehicle.remove(vehicle)
                log.info(f"{vehicle} rented.")
                print(f"{vehicle} rented successfully.")
            else:
                log.error(f"{vehicle} is not available for rent.")
                raise ValueError(f"{vehicle} is not available for rent.")
        except ValueError as v:
            print(v)

    # Implement a method for returning a vehicle in the system
    def return_vehicle(self, vehicle):
        """
        Add returned vehicle in the system.

        Parameter:
        vehicle (string): Vehicle name which you want to return in the system.

        Return:
        string: A message vehicle returned successfully.
        """
        self.available_vehicle.add(vehicle)
        print(f"{vehicle} returned successfully.")
        log.info(f"{vehicle} returned.")

    # Implement a method for display all vehicles in the system
    def display_vehicles(self):
        """
        Display all vehicle available in the system.

        Parameter:
        None

        Return:
        string: A message with all vehicle names.
        """
        print("Available vehicles: ")
        for vehicle in self.available_vehicle:
            print(vehicle)
        log.info(f"Available vehicles: {self.available_vehicle}")


# Do an example useage:
# Create a vehicle rental system
raj = vehicleRental()

# Add vehicles
raj.add_vehicle("bike1")
raj.add_vehicle("bike2")
raj.add_vehicle("car1")
raj.add_vehicle("car2")

# Display vehicles
raj.display_vehicles()

# Rent a vehicle
raj.rent_vehicle("car2")

# Display vehicles
raj.display_vehicles()

# Return a vehicle
raj.return_vehicle("car2")

# Display vehicles
raj.display_vehicles()
```

- Make a class named `vehicleRental`.
- Make an attribute named `availabe_vehicle`.
- Make `add_vehicle`, `rent_vehicle`, `return_vehicle`, `display_vehicles` methods for perform such operations.
- Do all kind of logging and exception handling in the class.
- Create a vehicle rental system with calling the class `vehicleRental`. Store it in `raj` object.

### 4. Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.

```python
# Create a book class
class book:
    """
    Make a Book with title and Author name.
    """
    log.info("Making a book...")

    # Implement three attribute called title, author and available
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.available = True

class library:
    """
    A simple library with add books, borrow book, return book, and display all books feature.
    """
    log.info("Starting library...")

    # Implement an attribute called books
    def __init__(self):
        self.books = []

    # Implement a method for add a book in the library system
    def add_book(self, book):
        """
        Function for add a book in the library.

        Parameter:
        book (iterator): Adding a book in the library. Make it an iterator using book class.

        Return:
        string: A message that book added successfully in the library.
        """
        self.books.append(book)
        print(f"Added '{book.title}' by {book.author} to the library.")
        log.info(f"Added '{book.title}' by {book.author} to the library.")

    # Implement a method for borrow a book from the library system
    def borrow_book(self, title):
        """
        Borrow a book from the library.

        Parameter:
        title (string): Book name which you want to borrow.

        Return:
        string: A message that book borrowd successfully or not.
        """
        try:
            for book in self.books:
                if book.title == title and book.available:
                    book.available = False
                    print(f"'{book.title}' by {book.author} borrowed successfully.")
                    log.info(f"Borrowed '{book.title}' by {book.author}.")
                    return
            log.error(f"'{title}' is either not available or doesn't exist in the library")
            raise IndexError(f"'{title}' is either not available or doesn't exist in the library.")
        except IndexError as i:
            print(i)
    
    # Implement a method for returning a book in the library system
    def return_book(self, title):
        """
        Return a book in the library.

        Parameter:
        title (string): Book name which you want to return.

        Return:
        string: A message that book returned successfully or not.
        """
        try:
            for book in self.books:
                if book.title == title and not book.available:
                    book.available = True
                    print(f"'{book.title}' by {book.author} returned successfully.")
                    log.info(f"Returned '{book.title}' by {book.author}.")
                    return
            log.error(f"'{title}' was not borrowed or doesn't exist in the library")
            raise IndexError(f"'{title}' was not borrowed or doesn't exist in the library.")
        except IndexError as i:
            print(i)
    
    # Implement a method for display all book which is available in the library system
    def display(self):
        """
        Display all books available in the library.

        Parameter:
        None

        Return:
        string: A message with all books name and author name available in the library.
        """
        print("Available books: ")
        for book in self.books:
            if book.available:
                print(f"'{book.title}' by {book.author}")
        
        log.info("Available books: ")
        for book in self.books:
            if book.available:
                log.info(f"'{book.title}' by {book.author}")


# Do an example usage:
# Create a library
lib = library()

# Add books
book1 = book("Let us C", "Yashavant Kanetkar")
book2 = book("Python for dummies", "Stef Maruch")
book3 = book("Rich Dad Poor Dad", "Robert Kiyosaki")
lib.add_book(book1)
lib.add_book(book2)
lib.add_book(book3)

# Display books
lib.display()

# Borrow two books
lib.borrow_book("Let us C")
lib.borrow_book("Python for dummies")

# Display books
lib.display()

# Return a book
lib.return_book("Let us C")

# Display books
lib.display()
```

- Make a class named `book` which take book name and author name and make an iterator.
- Make a class named `library`.
- Make an attribute named `books` for take a record of all books.
- Make `add_book`, `borrow_book`, `return_book`, `display` methods for perform such operations.
- Do all kind of logging and exception handling in the class.
- Create a library with calling the class `library`. Store it in `lib` object.
- Make book iterators by calling `book` class by passing it book title and the author name.

### 5. Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.

```python
# Create a product class
class product:
    """
    Make a product with id, name and quantity.
    """
    log.info("Making a product....")

    # Implement three attribute called product_id, name and quantity
    def __init__(self, product_id, name, quantity):
        self.product_id = product_id
        self.name = name
        self.quantity = quantity

# Create an inventory class
class inventory:
    """
    A simple inventory with add product, update quantity, and display all product feature.
    """
    log.info("Starting inventory....")

    # Implement an attribute called products
    def __init__(self):
        self.products = []

    # Implement a method for add a product in the inventory
    def add_product(self, product):
        """
        Function for add a product in the inventory.

        Parameter:
        product (iterator): Adding a product in the inventory . Make it an iterator using product class.

        Return:
        string: A message that product added successfully in the inventory.
        """
        self.products.append(product)
        print(f"Added {product.quantity} '{product.name}' (ID: {product.product_id}) to the inventory.")
        log.info(f"Added {product.quantity} '{product.name}' (ID: {product.product_id}) to the inventory.")

    # Implement a method for update quantity of a product in the inventory
    def update_quantity(self, product_id, new_quantity):
        """
        Update quantity of a product.

        Parameter:
        product_id (int): Product id of a product who's quantity to be updated.
        new_quantity (int): New product quantity of the product.

        Return:
        string: A message that product's new quantity updates or not.
        """
        try:
            for product in self.products:
                if product.product_id == product_id:
                    product.quantity = new_quantity
                    print(f"Updated quantity of '{product.name}' (ID: {product_id}) to {new_quantity}.")
                    log.info(f"Updated quantity of '{product.name}' (ID: {product_id}) to {new_quantity}.")
                    return
            log.error(f"Product with ID '{product_id}' doesn't exist in the inventory.")
            raise IndexError(f"Product with ID '{product_id}' doesn't exist in the inventory.")\
        
        except IndexError as i:
            print(i)

    # Implement a method for display all product id, names, and quantity available in the inventory
    def display(self):
        """
        Display all products available in the inventory.

        Parameter:
        None

        Return:
        string: A message with all product id, names and quantity available in the inventory.
        """
        print("Available Products: ")
        for product in self.products:    
            print(f"ID: {product.product_id}, Name: {product.name}, Quantity: {product.quantity}")
        
        log.info("Available Products: ")
        for product in self.products:
            log.info(f"ID: {product.product_id}, Name: {product.name}, Quantity: {product.quantity}")


# Do an example usage:
# Create an inventory
store = inventory()

# Add products
product1 = product(1201, "Laptop", 50)
product2 = product(1202, "Keyboard", 95)
product3 = product(1203, "Monitor", 20)
product4 = product(1204, "Mouse", 85)
store.add_product(product1)
store.add_product(product2)
store.add_product(product3)
store.add_product(product4)

# Display products
store.display()

# Update quantity
store.update_quantity(1203, 50)
store.update_quantity(1214, 68)
store.update_quantity(1201, 75)

# Display products
store.display()
```

### 6. Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.

```python
# Create a shape class
class shape:
    """
    Detect a shape using length, width and height.

    Parameter:
    length (int or float): Required. Length of the shape.
    width (int or float): Optional. Width of the shape.
    height (int or float): Optional. Height of the shape.

    Return:
    None
    """
    log.info("Shape detecting....")
    
    def __init__(self, length=0, width=None, height=None):
        self.length = length
        self.width = width
        self.height = height
    
    # Implement a method for calculate area of the shape
    def calculate_area(self):
        """
        Calculate area of the shape using length or lenght and width or length and height or length, height and width.

        Parameter:
        None

        Return:
        string: A maeeage that area of the shape or error if provided input (length, width or height) is wrong.
        """
        try:
            if self.width is None and self.height is None:
                area = self.length * self.length
                log.info(f"Area of a square with side {self.length}: {area}")
                print(f"Area of a square with side {self.length}: {area}")
            
            elif self.width is not None and self.height is None:
                area = self.length * self.width
                log.info(f"Area of a rectangle with length {self.length} and width {self.width}: {area}")
                print(f"Area of a rectangle with length {self.length} and width {self.width}: {area}")
            
            elif self.width is None and self.height is not None:
                area = (self.length * self.height) / 2
                log.info(f"Area of a triangle with length {self.length} and height {self.height}: {area}")
                print(f"Area of a triangle with length {self.length} and height {self.height}: {area}")
            
            elif self.width is not None and self.height is not None:
                area = 2 * (self.length*self.width + self.width*self.height + self.height*self.length)
                log.info(f"Surface area of cuboid with length {self.length}, width {self.width}, and height {self.height}: {area}")
                print(f"Surface area of cuboid with length {self.length}, width {self.width}, and height {self.height}: {area}")
            
            else:
                raise ValueError("Invalid dimensions provided.")
        except Exception as e:
            log.error(e)
            print(e)
    
    # Implement a method for calculate perimeter of the shape
    def calculate_perimeter(self):
        """
        Calculate perimeter of the shape using length or lenght and width or length and height or length, height and width.

        Parameter:
        None

        Return:
        string: A maeeage that perimeter of the shape or error if provided input (length, width or height) is wrong.
        """
        try:
            if self.width is None and self.height is None:
                perimeter = self.length * 4
                log.info(f"Perimeter of a square with side {self.length}: {perimeter}")
                print(f"Perimeter of a square with side {self.length}: {perimeter}")
            
            elif self.width is not None and self.height is None:
                perimeter = 2 * (self.length + self.width)
                log.info(f"Perimeter of a rectangle with length {self.length} and width {self.width}: {perimeter}")
                print(f"Perimeter of a rectangle with length {self.length} and width {self.width}: {perimeter}")
            
            else:
                raise ValueError("Invalid dimensions provided.")
        except Exception as e:
            log.error(e)
            print(e)
    

# Do an example usage
# Create a shape
shape1 = shape(6)
shape1.calculate_area()
shape1.calculate_perimeter()

# Create another shape
shape2 = shape(6, 4)
shape2.calculate_area()
shape2.calculate_perimeter()

# Create another two shapes
shape3 = shape(length=7, height=5)
shape3.calculate_area()
shape3.calculate_perimeter()

shape4 = shape(2, 4, 6)
shape4.calculate_area()
shape4.calculate_perimeter()
```

### 7. Student Management Create a class representing a student with attributes like student ID, name, and grades. Implement methods to calculate the average grade and display student details.

```python
# Create a student class
class student:
    """
    Make a student.

    Parameter:
    student_id (int): ID of a student.
    name (string): Name of the student.
    grades (list): A list of all grades of the student.

    Return:
    iterator: An student iterator.
    """
    log.info("Starting of student create.....")
    
    def __init__(self, student_id, name, grades=None):
        self.student_id = student_id
        self.name = name
        self.grades = grades if grades is not None else []

    # Implement a method called add_grade for adding grade in the grades list
    def add_grade(self, grade):
        """
        Add a grade in the list of grades.

        Parameter:
        grade (int or float): A grade which the student get and you want to store it.

        Return:
        string: A message that the grade added successfully in the grades list or not.
        """
        try:
            if isinstance(grade, (int, float)):
                self.grades.append(grade)
                print(f"Grade {grade} added for {self.name}")
                log.info(f"Grade {grade} added for {self.name}")
            else:
                raise ValueError("Grade should be a number.")
        except Exception as e:
            print(e)
            log.error(e)

    # Implement a method called average_grade for calculate average of all grades
    def average_grade(self):
        """
        Calculate average of grades.

        Parameter:
        None

        Return:
        string: A message that average grade of the student.
        """
        try:
            if len(self.grades) == 0:
                raise ValueError("No grades available to calculate average.")
            average_grade = sum(self.grades) / len(self.grades)
            print(f"Average grade for {self.name}: {average_grade}")
            log.info(f"Average grade for {self.name}: {average_grade}")
        except Exception as e:
            print(e)
            log.error(e)

    # Implement a method for display all details.
    def display(self):
        """
        Display all details about all students.

        Parameter:
        None

        Return:
        string: A message with all student details.
        """
        try:
            print(f"Student ID: {self.student_id}, Name: {self.name}, Grades: {self.grades}")
            log.info(f"Student ID: {self.student_id}, Name: {self.name}, Grades: {self.grades}")
        except Exception as e:
            print(e)
            log.error(e)


# Do an example usage:
# Make two students
student1 = student(12, "Dev", [20, 36, 42])
student2 = student(15, "Raj", [28, 35, 47])

# Display the students
student1.display()
student2.display()

# Add grades for the students
student1.add_grade(38)
student2.add_grade(49)

# Display the students
student1.display()
student2.display()

# Display their average number
student1.average_grade()
student2.average_grade()
```

- Make a class named `student` which take three attributes (student_id, name, grades).
- Make `add_grade`, `average_grade`, and `display` methods for perform such operations.
- Do all kind of logging and exception handling.
- Create some example students by calling `student` class. Store it in verious object (student1, student2).

```python
# Create a socialmedia class
class socialmedia:
    """
    Make a social media handler.

    Parameter:
    username (string): Take a username when creating a account.

    Return:
    iterator: A socialmedia iterator.
    """
    log.info("Starting of social media.....")
    
    def __init__(self, username):
        self.username = username
        self.posts = []

    # Implement a method for adding a post in social media
    def add_post(self, post):
        """
        Add a post in social media.

        Parameter:
        post (string): A post which user want to deliver.

        Return:
        string: Successfull or error message for the user.
        """
        try:
            if not isinstance(post, str):
                raise ValueError("Post content must be in string format.")
            
            self.posts.append(post)
            print(f"Added a new post: {post}")
            log.info(f"Added a new post: {post}")
        except ValueError as v:
            print(v)
            log.info(v)
    
    # Implement a method for display the post
    def display(self):
        """
        Display post(s) posted by user.

        Parameter:
        None

        Return:
        string: Post(s) or no post by user.
        """
        if self.posts:
            print(f"Posts by {self.username}:")
            log.info(f"Posts by {self.username}:")
            log.info(self.posts)
            for i, post in enumerate(self.posts, start=1):
                print(f"Post{i}: {post}")
        else:
            print(f"No post found for {self.username}")
            log.info(f"No post found for {self.username}")

    # Implement a method for searching a post
    def search_post(self, keyword):
        """
        Search a post using a keyword.

        Parameter:
        keyword (string): A keyword which user write in the post.

        Return:
        string: All the posts containing the keyword.
        """
        found_posts = []
        log.info(f"Entered keyword: {keyword}")

        try:
            if not isinstance(keyword, str):
                raise ValueError("Keyword must be a string.")
            
            for post in self.posts:
                if keyword.lower() in post.lower():
                    found_posts.append(post)

            if found_posts:
                print(f"Posts containing '{keyword}' by {self.username}")
                log.info(f"Posts containing '{keyword}' by {self.username}")
                log.info(found_posts)
                for i, post in enumerate(found_posts, start=1):
                    print(f"Post{i}: {post}")
            
            else:
                print(f"No posts containing '{keyword}' found for {self.username}")
                log.info(f"No posts containing '{keyword}' found for {self.username}")
        
        except ValueError as v:
            print(v)
            log.error(v)


# Do an example usage
# Make a profile
profile1 = socialmedia("Rajdip Das")

# Add some post
profile1.add_post("Hello everyone!! It's my first post.")
profile1.add_post("It is a beautiful day!!")
profile1.add_post("The flower is so beautiful.")

# View the posts
profile1.display()

# Search posts by a keyword
profile1.search_post("beautiful")
```

### 10. ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.

```python
# Create a todolist class
class todolist:
    """
    Make a todo list.

    Parameter:
    None

    Return:
    iterator: A todolist iterator.
    """
    log.info("Starting of Todolist....")

    def __init__(self):
        self.tasks = {}

    # Implement a method for add a task in todo list
    def add_task(self, task, due_date):
        """
        Add a task in ToDo List.

        Parameter:
        task (string): User's task which user want to add in todo list.
        due_date (string): The last date for which the task have to be done.
        
        Return:
        string: Successful or error message of the task added in todo list.
        """
        try:
            if not isinstance(task, str) or not isinstance(due_date, str):
                raise ValueError("Task and due date must be strings.")
            
            self.tasks[task] = due_date
            print(f"Added Task: {task}, Due: {due_date}")
            log.info(f"Added Task: {task}, Due: {due_date}")
        except ValueError as v:
            print(v)
            log.error(v)
    
    # Implement a method for mark a task as completed
    def mark_task(self, task):
        """
        Mark a task as completed.

        Parameter:
        task (string): Task which is completed by the user.

        Return:
        string: Successful of error message about the task marked as completed.
        """
        try:
            if task not in self.tasks:
                raise KeyError(f"Taks '{task}' not found.")
            
            del self.tasks[task]
            print(f"Marked task '{task}' as completed.")
            log.info(f"Marked task '{task}' as completed.")
        
        except KeyError as k:
            print(k)
            log.error(k)

    # Implement a method for display all pending tasks
    def display(self):
        """
        Display all pending tasks.

        Parameter:
        None

        Return:
        string: State of the pending tasks.
        """
        if self.tasks:
            print("Pending tasks:")
            log.info("Pending tasks:")
            log.info(self.tasks.items())
            for task, due_date in self.tasks.items():
                print(f"Task: {task} - Due: {due_date}")
        else:
            print("No pending tasks found.")
            log.info("No pending tasks found.")


# Do an example usage
# Make a todo list
todo = todolist()

# Add some task in todo list
todo.add_task("Finish assignment", "30-12-2023")
todo.add_task("Go to gym", "29-12-2023")
todo.add_task("Go to market", "29-12-2023")

# Display all pending tasks
todo.display()

# Mark a task as done
todo.mark_task("Go to market")

# Display all pending tasks
todo.display()
```

- Make `add_task`, `mark_task`, `display` methods for perform such operations.
- Do all kind of logging and exception handling.
- Create an example todo list by calling `todolist` class. Store it in `todo` object.
- Perform add task, mark as complete a task, display all pending tasks by calling such methods using the object and print prompt for user.
