```python
plt.plot(x, np.sin(x - 0), color='blue')
plt.ylim(-1.5, 1.5)
plt.xlim(0, 15)
plt.grid()
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("This is sin plot before real dataset execution to understand basic syntax")
plt.axis('equal')
plt.show()
```

```python
plt.plot(x, np.sin(x - 0), color='blue', label='Sinx')
plt.plot(x, np.cos(x - 0), color='red', label='Cosx')
plt.ylim(-1.5, 1.5)
plt.xlim(0, 15)
plt.grid()
plt.title("This is sin plot before real dataset execution to understand basic syntax")
plt.xlabel("x axis")
plt.ylabel("y axis")
# plt.axis('equal')
plt.legend()
plt.show()
```

```python
# Pie chart

langs = ['C', 'Python', 'Java']
students = [20, 100, 40]

plt.pie(students, labels=langs, autopct='%1.1f%%')
plt.title("Pie chart")
plt.show()
```

```python
my_explode = (0.0, 0.1, 0.1)

langs = ['C', 'Python', 'Java']
students = [20, 100, 40]

plt.pie(students, labels=langs, autopct='%1.1f%%', explode=my_explode, shadow=True)
plt.title('Pie chart')
plt.show()
```

```python
# 3D plot

x = np.random.rand(20)
y = np.random.rand(20)
z = np.random.rand(20)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x, y, z, c="#FF5733", alpha=.9)
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```
