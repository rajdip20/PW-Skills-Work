### 1. Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.

```python
# Make bank class
class bank:
    """
    A mini banking system with deposit and withdraw feature.
    """
    log.info("Start of Banking System....")

    # Creating three attributes
    def __init__(self, accnum, name, balance=0):
        self.account_num = accnum
        self.holder_name = name
        self.balance = balance

    # Implement of deposit method
    def deposit(self, amount):
        """
        Function for deposit amount.

        Parameter:
        amount (int or float): The amount to be deposited.

        Return:
        int or float: The Total amount in the account of account holder.
        """
        try:
            if amount > 0:
                self.balance += amount
                log.info(f"Amount {amount} deposit successfully. Depositer current balance {self.balance}.")

                print(f"Amount {amount} deposit successfully. Your current balance {self.balance}.")
            else:
                log.error("Typed amount is less than 1.")
                raise ValueError("Deposit amount should be greater than 0.")
        
        except ValueError as v:
            print(v)
    
    # Implement of withdraw method
    def withdraw(self, amount):
        """
        Function for withdraw amount.

        Parameter:
        amount (int or float): The amount to be withdrawl.

        Return:
        int or float: The Total amount in the account of account holder.
        """
        try:
            if amount > 0:
                if self.balance >= amount:
                    self.balance -= amount
                    log.info(f"Take {amount}. Current balance is {self.balance}.")
                    
                    print(f"Take your amount {amount}. Your current balance {self.balance}.")
                
                else:
                    log.error("Typed amount greater than available balance.")
                    raise ValueError("Insufficient funds.")
            else:
                log.error("Typed amount is less than 1.")
                raise ValueError("Withdraw amount must be greater than 0.")
        
        except ValueError as v:
            print(v)


# Do an example usage:
# Create a bank account
acc = bank("123456789", "Rajdip Das", 10000)

# Deposit money
acc.deposit(500)

# Withdraw money
acc.withdraw(2000)  # This will log an error due to insufficient funds

# Another deposit
acc.deposit(100)
```

- Make a class named `bank`.
- Make 3 attribute named `account_num`, `holder_name`, `balance`.
- Make `deposit` and `withdraw` methods for perform such operations.
- Do all kinds of logging and exception handling in the class.
- Create a bank account with calling the class `bank` and pass arguments like, account number, name, and initial balance. Store it in `acc` object.
- Perform deposit and withdraw operations calling such methods using the object and print the deposit/withdraw balance and current balance.

### 2. Employee Management Create a class representing an employee with attributes like employee ID, name, and salary. Implement methods to calculate the yearly bonus and display employee details.

```python
# Create a employee class
class employee:
    """
    A employee management system. 
    
    Calculate yearly salary and display employee details.
    """
    log.info("Start of employee management system...")

    # Implement three attributes like emp_id, name and salary
    def __init__(self, emp_id, name, salary):
        self.emp_id = emp_id
        self.name = name
        self.salary = salary

    # Implement a method for calculate yearly bonus
    def calculate_yearly_bonus(self, percent):
        """
        Calculate yearly salary of the employee.

        Parameter:
        percent (int or float): How much yearly bonus percentage the employee got.

        Return:
        int or float: How much bonus amount the employee got.
        """
        try:
            if percent >= 0:
                self.bonus = (percent / 100) * self.salary
                log.info(f"Calculated yearly bonus for {self.salary}: {self.bonus}")
                return self.bonus
        
            else:
                log.error("Entered bonus percentage should be a positive number.")
                raise ValueError("Bonus percentage should be a positive number.")
            
        except ValueError as v:
            print(v)

    # Implement a method for display a employee details
    def display(self):
        """
        Display all the employee details.

        Parameter:
        None.

        Return:
        string: All details of an employee.
        """
        log.info("Show the employee details...")
        log.info(f"Employee ID: {self.emp_id}, Name: {self.name}, Salary: {self.salary}, bonus: {self.bonus}")
        print(f"Employee ID: {self.emp_id}, Name: {self.name}, Salary: {self.salary}, bonus: {self.bonus}")


# Do an example usage:
# Create an employee
emp = employee(1234, "Rajdip Das", 40000)

# Calculate yearly bonus and display it
bonus_amount = emp.calculate_yearly_bonus(20)
print("The bonus amount:", bonus_amount)

# Display employee details
emp.display()
```

- Make a class named `employee`.
- Make 3 attribute named `emp_id`, `name`, `salary`.
- Make `calculate_yearly_bonus` and `display` methods for perform such operations.
- Do all kind of logging and exception handling in the class.
- Create an employee with calling the class `employee` and pass arguments like, employee id, name and salary. Store it in `emp` object.
- Perform yearly bonus calculation and display the information of the employee operations by calling such methods using the object and print the yearly bonus and employee details.

### 3. Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle, return a vehicle, and display available vehicles.

```python
# Create a vehicleRental class
class vehicleRental:
    """
    A simple vehicle rental system.
    """
    log.info("Starting vehicle rental system....")

    # Implement an attribute called available_vehicle
    def __init__(self):
        self.available_vehicle = set()

    # Implement a method for add vehicle in the system
    def add_vehicle(self, vehicle):
        """
        Add a vehicle in the system.

        Parameter:
        vehicle (string): Vehicle name which you want to added in system.

        Return:
        string: A message vehicle added successfully in the system.
        """
        self.available_vehicle.add(vehicle)
        print(f"{vehicle} added successfully.")
        log.info(f"{vehicle} added to available vehicles.")

    # Implement a method for rent a vehicle from the system
    def rent_vehicle(self, vehicle):
        """
        Rent a vehicle from the system.

        Parameter:
        vehicle (string): Vehicle name which you want to rent from the system.

        Return:
        string: A message vehicle rent successfully or not.
        """
        try:
            if vehicle in self.available_vehicle:
                self.available_vehicle.remove(vehicle)
                log.info(f"{vehicle} rented.")
                print(f"{vehicle} rented successfully.")
            else:
                log.error(f"{vehicle} is not available for rent.")
                raise ValueError(f"{vehicle} is not available for rent.")
        except ValueError as v:
            print(v)

    # Implement a method for returning a vehicle in the system
    def return_vehicle(self, vehicle):
        """
        Add returned vehicle in the system.

        Parameter:
        vehicle (string): Vehicle name which you want to return in the system.

        Return:
        string: A message vehicle returned successfully.
        """
        self.available_vehicle.add(vehicle)
        print(f"{vehicle} returned successfully.")
        log.info(f"{vehicle} returned.")

    # Implement a method for display all vehicles in the system
    def display_vehicles(self):
        """
        Display all vehicle available in the system.

        Parameter:
        None

        Return:
        string: A message with all vehicle names.
        """
        print("Available vehicles: ")
        for vehicle in self.available_vehicle:
            print(vehicle)
        log.info(f"Available vehicles: {self.available_vehicle}")


# Do an example useage:
# Create a vehicle rental system
raj = vehicleRental()

# Add vehicles
raj.add_vehicle("bike1")
raj.add_vehicle("bike2")
raj.add_vehicle("car1")
raj.add_vehicle("car2")

# Display vehicles
raj.display_vehicles()

# Rent a vehicle
raj.rent_vehicle("car2")

# Display vehicles
raj.display_vehicles()

# Return a vehicle
raj.return_vehicle("car2")

# Display vehicles
raj.display_vehicles()
```

- Make a class named `vehicleRental`.
- Make an attribute named `availabe_vehicle`.
- Make `add_vehicle`, `rent_vehicle`, `return_vehicle`, `display_vehicles` methods for perform such operations.
- Do all kind of logging and exception handling in the class.
- Create a vehicle rental system with calling the class `vehicleRental`. Store it in `raj` object.
- Perform add vehicle, rent vehicle, return vehicle and display vehicles operations by calling such methods using the object and print a prompt for user.

### 4. Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.

```python
# Create a book class
class book:
    """
    Make a Book with title and Author name.
    """
    log.info("Making a book...")

    # Implement three attribute called title, author and available
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.available = True

class library:
    """
    A simple library with add books, borrow book, return book, and display all books feature.
    """
    log.info("Starting library...")

    # Implement an attribute called books
    def __init__(self):
        self.books = []

    # Implement a method for add a book in the library system
    def add_book(self, book):
        """
        Function for add a book in the library.

        Parameter:
        book (iterator): Adding a book in the library. Make it an iterator using book class.

        Return:
        string: A message that book added successfully in the library.
        """
        self.books.append(book)
        print(f"Added '{book.title}' by {book.author} to the library.")
        log.info(f"Added '{book.title}' by {book.author} to the library.")

    # Implement a method for borrow a book from the library system
    def borrow_book(self, title):
        """
        Borrow a book from the library.

        Parameter:
        title (string): Book name which you want to borrow.

        Return:
        string: A message that book borrowd successfully or not.
        """
        try:
            for book in self.books:
                if book.title == title and book.available:
                    book.available = False
                    print(f"'{book.title}' by {book.author} borrowed successfully.")
                    log.info(f"Borrowed '{book.title}' by {book.author}.")
                    return
            log.error(f"'{title}' is either not available or doesn't exist in the library")
            raise IndexError(f"'{title}' is either not available or doesn't exist in the library.")
        except IndexError as i:
            print(i)
    
    # Implement a method for returning a book in the library system
    def return_book(self, title):
        """
        Return a book in the library.

        Parameter:
        title (string): Book name which you want to return.

        Return:
        string: A message that book returned successfully or not.
        """
        try:
            for book in self.books:
                if book.title == title and not book.available:
                    book.available = True
                    print(f"'{book.title}' by {book.author} returned successfully.")
                    log.info(f"Returned '{book.title}' by {book.author}.")
                    return
            log.error(f"'{title}' was not borrowed or doesn't exist in the library")
            raise IndexError(f"'{title}' was not borrowed or doesn't exist in the library.")
        except IndexError as i:
            print(i)
    
    # Implement a method for display all book which is available in the library system
    def display(self):
        """
        Display all books available in the library.

        Parameter:
        None

        Return:
        string: A message with all books name and author name available in the library.
        """
        print("Available books: ")
        for book in self.books:
            if book.available:
                print(f"'{book.title}' by {book.author}")
        
        log.info("Available books: ")
        for book in self.books:
            if book.available:
                log.info(f"'{book.title}' by {book.author}")


# Do an example usage:
# Create a library
lib = library()

# Add books
book1 = book("Let us C", "Yashavant Kanetkar")
book2 = book("Python for dummies", "Stef Maruch")
book3 = book("Rich Dad Poor Dad", "Robert Kiyosaki")
lib.add_book(book1)
lib.add_book(book2)
lib.add_book(book3)

# Display books
lib.display()

# Borrow two books
lib.borrow_book("Let us C")
lib.borrow_book("Python for dummies")

# Display books
lib.display()

# Return a book
lib.return_book("Let us C")

# Display books
lib.display()
```

- Make a class named `book` which take book name and author name and make an iterator.
- Make a class named `library`.
- Make an attribute named `books` for take a record of all books.
- Make `add_book`, `borrow_book`, `return_book`, `display` methods for perform such operations.
- Do all kind of logging and exception handling in the class.
- Create a library with calling the class `library`. Store it in `lib` object.
- Make book iterators by calling `book` class by passing it book title and the author name.
- Perform add book, borrow book, return book, and display all available books operations by calling such methods using the object and print a prompt for user.

### 5. Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.

```python
# Create a product class
class product:
    """
    Make a product with id, name and quantity.
    """
    log.info("Making a product....")

    # Implement three attribute called product_id, name and quantity
    def __init__(self, product_id, name, quantity):
        self.product_id = product_id
        self.name = name
        self.quantity = quantity


class inventory:
    """
    A simple inventory with add product, update quantity, and display all product feature.
    """
    log.info("Starting inventory....")

    # Implement an attribute called products
    def __init__(self):
        self.products = []

    # Implement a method for add a product in the inventory
    def add_product(self, product):
        """
        Function for add a product in the inventory.

        Parameter:
        product (iterator): Adding a product in the inventory . Make it an iterator using product class.

        Return:
        string: A message that product added successfully in the inventory.
        """
        self.products.append(product)
        print(f"Added {product.quantity} '{product.name}' (ID: {product.product_id}) to the inventory.")
        log.info(f"Added {product.quantity} '{product.name}' (ID: {product.product_id}) to the inventory.")

    # Implement a method for update quantity of a product in the inventory
    def update_quantity(self, product_id, new_quantity):
        """
        Update quantity of a product.

        Parameter:
        product_id (int): Product id of a product who's quantity to be updated.
        new_quantity (int): New product quantity of the product.

        Return:
        string: A message that product's new quantity updates or not.
        """
        try:
            for product in self.products:
                if product.product_id == product_id:
                    product.quantity = new_quantity
                    print(f"Updated quantity of '{product.name}' (ID: {product_id}) to {new_quantity}.")
                    log.info(f"Updated quantity of '{product.name}' (ID: {product_id}) to {new_quantity}.")
                    return
            log.error(f"Product with ID '{product_id}' doesn't exist in the inventory.")
            raise IndexError(f"Product with ID '{product_id}' doesn't exist in the inventory.")\
        
        except IndexError as i:
            print(i)

    # Implement a method for display all product id, names, and quantity available in the inventory
    def display(self):
        """
        Display all products available in the inventory.

        Parameter:
        None

        Return:
        string: A message with all product id, names and quantity available in the inventory.
        """
        print("Available Products: ")
        for product in self.products:    
            print(f"ID: {product.product_id}, Name: {product.name}, Quantity: {product.quantity}")
        
        log.info("Available Products: ")
        for product in self.products:
            log.info(f"ID: {product.product_id}, Name: {product.name}, Quantity: {product.quantity}")


# Do an example usage:
# Create an inventory
store = inventory()

# Add products
product1 = product(1201, "Laptop", 50)
product2 = product(1202, "Keyboard", 95)
product3 = product(1203, "Monitor", 20)
product4 = product(1204, "Mouse", 85)
store.add_product(product1)
store.add_product(product2)
store.add_product(product3)
store.add_product(product4)

# Display products
store.display()

# Update quantity
store.update_quantity(1203, 50)
store.update_quantity(1214, 68)
store.update_quantity(1201, 75)

# Display products
store.display()
```

- Make a class named `product` which take product id, name, and quantity. Then make these an iterator.
- Make a class named `inventory`.
- Make an attribute named `products` for take a record of all products.
- Make `add_product`, `update_quantity`, `display` methods for perform such operations.
- Do all kind of logging and exception handling in the class.
- Create an inventory by calling the class `inventory`. Store it in `store` object.
- Make product iterators by calling `product` class by passing it product id, name, and the quantity.
- Perform add product, update quantity, and display all available products operations by calling such methods using the object and print a prompt for user.
